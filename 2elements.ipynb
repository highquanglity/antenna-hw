```python
import numpy as np  
import matplotlib.pyplot as plt
import math
%matplotlib inline
e0=1
theta=np.linspace(0,2*np.pi,1000)
```

# **1) 2 isotropic point sources of same amplitude and phase, d= λ** 



\begin{align}
    E =&  E_{0}e^{-j\beta(\frac{d}{2}\cosγ)} +  E_{0}e^{j\beta(\frac{d}{2}\cosγ)} \\  
     =&  E_{0}e^{-j(\frac{2π}{λ})(\frac{λ}{2}\cosγ)} +   E_{0}e^{j(\frac{2π}{λ})(\frac{λ}{2}\cosγ)}\\
     =&  E_{0}e^{-j(π\cosγ)} +  E_{0}e^{j(π\cosγ)} \\
     =&  2E_{0}\cos{(\pi\cos{γ})}
    \end{align}

```python
curve=abs(2*e0*np.cos((np.pi)*np.cos(theta)))
fig=plt.figure(figsize=(8,8))
ax1=plt.subplot(111,polar=True)
for t in theta:
    ax1.plot(theta,curve,color="black")
plt.show()
```

# **2) 2 isotropic point  sources of same amplitude, 45º phase difference, d= λ**

\begin{align}
    E =&  E_{0}e^{-j\beta(\frac{d}{2}\cosγ)-\frac{\pi}{8}} +  E_{0}e^{j\beta(\frac{d}{2}\cosγ)+\frac{\pi}{8}} \\  
     =&  E_{0}e^{-j(\frac{2π}{λ})(\frac{λ}{2}\cosγ)-\frac{\pi}{8}} +   E_{0}e^{j(\frac{2π}{λ})(\frac{λ}{2}\cosγ)+\frac{\pi}{8}}\\
     =&  E_{0}e^{-j(π\cosγ)-\frac{\pi}{8}} +  E_{0}e^{j(π\cosγ)+\frac{\pi}{8}} \\
     =&  2E_{0}\cos{(\pi\cos{γ+\frac{\pi}{8}})}
    \end{align}

```python
curve=abs(2*e0*np.cos(np.pi*np.cos(theta)+np.pi/8))
fig=plt.figure(figsize=(8,8))
ax1=plt.subplot(111,polar=True)
for t in theta:
    ax1.plot(theta,curve,color='black')
plt.show()
```
